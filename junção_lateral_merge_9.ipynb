####    Merge - Parâmetros:
    Left e Right: São os parâmetros que indicam quais serão os dataframes que deverão ser combinados;
    how: 
        - Como a junção será feita:
            • left: Irá utilizar apenas a(s) chave(s) do DataFrame indicado no parâmetro "left", ou seja, 
            todos os dados desse DataFrame estarão presentes no novo DF;

            • right: Irá utilizar apenas a(s) chave(s) do DataFrame indicado no parâmetro "right", ou seja, 
            todos os dados desse DataFrame estarão presentes no novo DF;

            obs: Se os dois df's left e right estiverem sendo unidos por uma chave em comum e um dos dois parâmetros acima 
            forem aplicados, somente serão unidos os dadosque estiverem em comum usando como base os dados do dataframe 
            escolhido como parâmetro, seja left ou right;

            • inner: Realiza a interseção dessas duas variáveis de ligação (left e right), somente quando existirem nos dois 
            DataFrames é que irá existir o valor no novo DataFrame, caso exista em um e no outro não o valor será deletado do novo DF.

            • outer: Realiza a união de todas as informações, as informações dos dois dataframes vão aparecer, 
            quando tiver união, ficaram juntas e quando não tiver vão ficar soltas.

    on: 
        - Quando temos a variável de ligação igual nos 2 Dataframes, passamos essa variável para indicar que ela será 
        a chave de ligação da união dos DataFrames.

    Left_on e Right_on: 
        - Usados quando para quando a variável de ligação dos DataFrames estão diferentes, desta forma 
        indicamos dentro do parâmetro quais será(ão) a(s) chave(s) de ligação, podem ser 1 ou mais, esta chave de ligação será
        referente somente ao DF escolhido sendo left_on para chaves de ligação do DataFrame setado com left e right_on para as 
        chaves de ligação do DataFrame setado como right.

    Left_index e Right_index: 
        - Booleanos (True e False) indicando se usaremos o índice como variável do arquivo da esquerda 
        ou do arquivo da direita (Indicado para criação de DataFrames apartir de um Groupby).
