# Como transformar o arquivo Formato Json em Tabela:

json_normalize: 

        Irá pegar o valor dos dados json carregados e para cada conjunto de chave e valor dentro do arquivo ele irá transformar em uma coluna tendo a chave como nome da coluna e valor como valor da linha carregada. 
        
        Obs: caso o conjunto chave e valor seja seguido de outro conjunto chave e valor ele irá agregar as chaves da seguinte forma: 
        
        Json Carregado:
        
                {'normal':{'tipo':{'propriedade':123}}}

        Json tratado:

                |normal.tipo.propriedade |
                |                        |
                |123                     |

        Caso encontre uma 'variavel final' que pode ser um valor str, number, list, etc, o json_normalize irá definir esse dado como o valor da coluna.


        Sintaxe:
                
                dados = pd.json_normalize(data='arquivo_json_carregado', )


        Exemplo:
        
                Entrada:
                        dados = pd.read_json(path_or_buf = 'stage1.1/realestates.json', orient = 'columns')

                Saída:
                        >>> normal	highlights
                        >>> output	{'listings': [{'imovel': {'tipos': {'proprieda...	{'listings': [{'imovel': {'tipos': {'proprieda...

                Entrada:
                        dados = pd.json_normalize(data=dados.normal)
                        dados.head()

                Saída:
                        listings: [{'imovel': {'tipos': {'proprieda...

# Porque arquivo ainda ficou formatado no modelo de dicionario/json?

    Como sitado acima, a função json_normalize ao se deparar com um valor que pode ser considerado 'final' ela para a formatação do arquivo poís considera que aquele pode ser o valor final da coluna.

    Neste caso, ao aplicar o json_normalize no conjunto de dados 'dados' onde foi carregado o arquivo completo json, a função se deparou com o conjunto chave:valor onde a chave é o valor 'listings' e o valor uma lista [], logo o pandas assumiu que aquele seria o ponto final da tratativa para a função pd.json_normalize, retornando a tabela contendo somente uma coluna 'listings' com uma única linha contendo a lista com o restante dos dados!

# Como posso transformar a lista recebida na primeira linha em linhas de uma tabela?

    Podemos usar a atribuição direta da lista de arquivos json:

    Entrada:
        
        dados_normal_listings = pd.json_normalize(data=dados[listing].loc[0], sep='_')

    Saída:

    >>> Dataframe


    Obs: Existe também a opção de como definir até que nível de conjunto chave e valor o normalize irá tratar, para fazer isso é preciso passar o parametro desejado de niveis para 'max_level'. Ex:

    Entrada:
        
        dados_normal_listings = pd.json_normalize(data=dados[listing].loc[0], sep='_', max_level=2)

    Saída:

    >>> Dataframe

```python
dados_normal_listings = pd.json_normalize(data=dados['listings'].loc[0], sep='_')
dados_normal_listings.head()
```

```python
dados_normal_listings = pd.json_normalize(data=dados['listings'].loc[0], sep='_', max_level=2)
dados_normal_listings.head()
```

# Como fazer a transformação do json em tabela de forma direta?

Podemos fazer isso passando para a primeira tabela criada na leitura do json como 'data' e passando como record_path a lista que viria a ser a lista que seria tratada como visto na forma anterior:

```python
dados = pd.read_json(path_or_buf = 'stage1.1/realestates.json', orient = 'columns')

dados_normal_listings = pd.json_normalize(data=dados.normal, sep='_', record_path=['listings'])
dados_normal_listings.head()
```

```python
dados_higlights_listings = pd.json_normalize(data=dados.highlights, sep='_', record_path=['listings'])
dados_higlights_listings.head()
```
