```python
import pandas as pd
import numpy as np 
```

```python
dados = pd.read_json(path_or_buf = 'stage1.1/realestates.json', orient = 'columns')
dados_normal_listings = pd.json_normalize(data=dados.normal, sep='_', record_path=['listings'])
dados_normal_listings.head()
```

```python
dados_higlights_listings = pd.json_normalize(data=dados.highlights, sep='_', record_path=['listings'])
dados_higlights_listings
```

### Filtrar e criar um subconjunto de colunas no DataFrame:
    Função: pd.DataFrame.filter();

        -   Irá realizar a busca das linhas ou colunas especificadas dentro de seus paramêtros:


#### Buscando colunas desejadas:

```python
dados_normal_listings.filter(like='imovel_caracteristicas')
```

#### Resgatando somente o nome das colunas desejadas:

```python
colunas = dados_normal_listings.filter(like='imovel_caracteristicas')
colunas
```

• Agora que temos uma lista com os nomes das colunas que desejamos tratar, podemos usar essa lista para realizar uma iteração e assim conseguir aplir os filtros necessários em forma de loop:

#### Criando loop que use a lista de colunas como iterador:

```python
for coluna in colunas:
    print(coluna)
```

•   Com o iterador criado, podemos atribuir os filtros criados na aula 5 para as colunas contidas na lista 'colunas':

        - Podemos aplicar este filtro para os dois DataFrames: normal e highlights;

```python
for coluna in colunas: 
    dados_normal_listings[coluna] = dados_normal_listings[coluna].str[1:-1].str.replace("'","").str.split(',')
    dados_higlights_listings[coluna] = dados_higlights_listings[coluna].str[1:-1].str.replace("'","").str.split(',')
```

• Agora é possível realizar o resgate de informações acessando a posição do item dentro da lista da célula, o que anteriormente era impossível ja que a célula era um objeto, então,
ao acessar a posição dentro do objeto ele retornava um caracter.

• Como transformamos todas as celulas da coluna em lista, agora torna-se possível o resgate das informações completas por posição dentro da célula:

```python
dados_normal_listings.loc[0,'imovel_caracteristicas_propriedade'][0]
```

```python
dados_higlights_listings.loc[0,'imovel_caracteristicas_propriedade'][0]
```
